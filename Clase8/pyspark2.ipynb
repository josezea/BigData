```
file_location = "/FileStore/tables/s11_2019II.txt"
df = spark.read.option("header","true")\
.option("delimiter","|")\
    .option("inferSchema","true")\
.csv(file_location) 
display(df)
```

```
datos=df.select(['ESTU_TIPODOCUMENTO', 'ESTU_CONSECUTIVO', 'ESTU_PAIS_RESIDE', 'COLE_COD_DANE_ESTABLECIMIENTO','COLE_NOMBRE_ESTABLECIMIENTO', 'COLE_COD_MCPIO_UBICACION', 'COLE_COD_DEPTO_UBICACION', 'COLE_NATURALEZA', 'COLE_JORNADA','COLE_CALENDARIO','COLE_BILINGUE','PUNT_GLOBAL','PUNT_MATEMATICAS','PUNT_INGLES','PUNT_C_NATURALES',
                 'ESTU_NSE_INDIVIDUAL','ESTU_NSE_ESTABLECIMIENTO'])
datos.show()
```

```
# Eliminar columnas
datos=datos.drop('ESTU_PAIS_RESIDE')
```

```
# Realizar el filtro de los colegios de Bogotá 
datos.groupBy('COLE_COD_MCPIO_UBICACION').count().show()
```

```
df.schema['COLE_COD_MCPIO_UBICACION'].dataType
```

```
divipola.show()
```

```
divipola = divipola.withColumnRenamed("Código Municipio","COLE_COD_MCPIO_UBICACION") \
    .withColumnRenamed("Nombre Municipio","nom_mpio")
divipola.printSchema()
```

```
datos=datos.join(divipola, on='COLE_COD_MCPIO_UBICACION')
display(datos)
```

```
# Filtrar los colegios de Bogotá
datos_bta=datos[datos['COLE_COD_MCPIO_UBICACION']==11001]
display(datos_bta)
```

```
#Construir puntaje STEM
datos=datos.withColumn('puntaje_stem',datos.PUNT_INGLES+datos.PUNT_MATEMATICAS/2)
display(datos)
```

```
#Agregación: igual que en pandas
```

```
consulta = datos.groupby(['COLE_JORNADA']).agg({'COLE_JORNADA':'size','COLE_NATURALEZA':'mean','PUNT_GLOBAL':'mean'})
display(consulta)
```

```
consulta2 = datos.groupby(['COLE_JORNADA']).agg({'ESTU_CONSECUTIVO':'size','PUNT_GLOBAL':'mean','PUNT_MATEMATICAS':'max'})
display(consulta2)
```
